# The most used words (non-articles)

```python
import re
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
findings[:5]
```

```python
    d={}
    for word in findings:
        if word in d.keys():
            d[word] = d[word] + 1
        else:
            d[word] = 1
```

```python
d_list = [(value, key) for (key, value) in d.items()]
d_list = sorted(d_list, reverse=True)
```

# Sentiment Anyalysis: What is the most positive and most negative chapter?

### An example

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
scores = analyzer.polarity_scores("Hey, look how beautiful the trees are. I love them.")
```

```python
if scores["pos"] > scores ["neg"]:
    print("It is a positive text")
else:
    print("It is a negative text")
    
```

```python
analyzer.polarity(book)
```

### Chapters sentiment anyalysis

```python
import re
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
chapters = chapters[1:]
```

```python
for chapter in chapters:
    scores = analyzer.polarity_scores(chapter)
    print(scores)
```
